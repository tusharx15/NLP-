__TOKENISATION__

```python
import nltk
from indicnlp.tokenize import sentence_tokenize, indic_tokenize
from nltk.probability import FreqDist


#Tokenize sentences
sentences = sentence_tokenize.sentence_split(text, lang='hi')

# Tokenize words within each sentence
tokenized_sentences = [word for sentence in sentences for word in indic_tokenize.trivial_tokenize(sentence)]

# Creating a frequency distribution of tokens:
frequency_dist = FreqDist(tokenized_sentences)

#Print tokenized sentences
for hindi_tokens in tokenized_sentences:
    print(indic_tokenize.trivial_tokenize(hindi_tokens))

# Print the frequency distribution
print(frequency_dist)
```

__2.) FOREIGN LABNGUAGE-SPANISH__

```python
import PyPDF2

# Open the PDF file
pdf_file = open("spanishtext.pdf", "rb")

# Initialize PDF reader
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Extract text from each page
text = ""
for page in pdf_reader.pages:
    text += page.extract_text()

# Close the PDF file
pdf_file.close()
```

__TOKENISATION__

```python
import nltk

# Tokenize the Spanish text into words
spanish_words = nltk.word_tokenize(text, language='spanish')


freq = FreqDist(spanish_words)

# Print the tokenized words
print(spanish_words)


# Print the frequency distribution
print(freq)
```

# STEMMING THE TOKENS OF EACH LANGUAGE-

```python
import pandas as pd
import numpy as np
```

```python
# Convert the frequency distribution to a dictionary
freq_dict = dict(frequency_dist)

# Create dataframes
df_indian = pd.DataFrame(list(frequency_dist.items()), columns=["Token", "Frequency_Indian"])
df_foreign = pd.DataFrame(list(freq.items()), columns=["Token", "Frequency_Spanish"])

# Merge the dataframes using pandas concat
combined_df = pd.merge(df_indian, df_foreign, on="Token", how="outer")

# Fill NaN values with 0 (if a token is not found in one of the languages)
combined_df = combined_df.fillna(0)

# # Sort the combined dataframe by frequency (optional)
# combined_df = combined_df.sort_values(by=["Frequency_Indian", "Frequency_Espana"], ascending=False)

# Display the combined dataframe
print(combined_df)
```

```python
from nltk.stem import ISRIStemmer
from nltk.stem import SnowballStemmer


# Function to perform Hindi stemming 
def stem_hindi(word):
    hindi_stemmer = ISRIStemmer()
    word = hindi_stemmer.stem(word) 
    return word

# Function to perform Spanish stemming
def stem_spanish(word):
    stemmer = SnowballStemmer("spanish")
    return stemmer.stem(word)

# Apply stemming for Hindi and Spanish
df_indian["Stem_Indian"] = df_indian["Token"].apply(stem_hindi)
df_foreign["Stem_Spanish"] = df_foreign["Token"].apply(stem_spanish)

# Display the updated DataFrame
print(df_indian)

print(df_foreign)
```

```python
# Merge the dataframes using pandas merge
combined_df = pd.merge(df_indian, df_foreign, on="Token", how="outer")

# Fill NaN values with 0 (if a token is not found in one of the languages)
combined_df = combined_df.fillna(0)
```
