```python
str
words = str.split(' ')
words
```

```python
from collections import Counter
word_count = Counter(words)
print(word_count)
```

```python
counts = {item:words.count(item) for item in words}
```

```python
sorted_dict = dict(sorted(counts.items(), key=lambda x: x[1], reverse = True))
print(sorted_dict)
```

```python
import matplotlib.pylab as plt
```

```python
plt.bar(range(len(sorted_dict)), sorted_dict.values(), align='center')
plt.xticks(range(len(sorted_dict)),sorted_dict.keys())

plt.show()
```
