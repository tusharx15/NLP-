```python
import nltk
import pandas as pd
```

```python
# Download NLTK resources for part-of-speech tagging
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
# Sample statements
statements = [
    "Football, or soccer, is the world's most popular sport with billions of fans.",
    "The modern version of football originated in England in the 19th century.",
    "The FIFA World Cup is the most prestigious international football tournament, held every four years.",
    "Lionel Messi and Cristiano Ronaldo are widely regarded as two of the greatest football players of all time.",
    "The offside rule prevents attacking players from positioning behind the last defender when the ball is played.",
    "The football field, or pitch, is typically rectangular, and each team consists of 11 players.",
    "The UEFA Champions League is an annual club competition featuring the best European teams.",
    "Yellow and red cards are used to discipline players for misconduct, with a red card resulting in a player being sent off.",
    "VAR (Video Assistant Referee) technology assists referees in making more accurate decisions.",
    "A hat-trick in football refers to a player scoring three goals in a single game."
]
```

```python
# Perform part-of-speech tagging for each statement
pos_tags_statements = [nltk.pos_tag(nltk.word_tokenize(statement)) for statement in statements]
```

```python
# Display statements with corresponding POS tags
for i, (statement, pos_tags) in enumerate(zip(statements, pos_tags_statements), start=1):
    print(f"Statement {i}: {statement}")
    for tag, words in nltk.FreqDist(pos[1] for pos in pos_tags).items():
        print(f"{tag}: {', '.join(word for word, t in pos_tags if t == tag)}")
    print()
```

# WORD EMBEDDING

```python
# Importing the neccessry libraries-
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
import string
```

```python
# Download NLTK resources for stopwords and punkt
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Tokenization, removing stopwords and punctuation
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)
```

```python
tokenized_statements = []
for statement in statements:
    words = word_tokenize(statement)
    filtered_words = [word.lower() for word in words if word.lower() not in stop_words and word.lower() not in punctuation]
    tokenized_statements.append(filtered_words)
```

```python
# Word Embedding using Word2Vec
model = Word2Vec(tokenized_statements, vector_size=100, window=5, min_count=1, workers=4)
```

```python
# Example: Get the vector representation of a word
word_vector = model.wv['technology']
```

```python
# Display the vector representation of the word 'generative'
print("Vector representation of 'technology':", word_vector)
```

```python
import pandas as pd

# Create a DataFrame to store similarity scores
similarity_df = pd.DataFrame(columns=['Token'] + model.wv.index_to_key)
```

```python
# Iterate through each token in the Word2Vec model vocabulary
for token1 in model.wv.index_to_key:
    # Initialize a list to store similarity scores for the current token
    similarity_scores = [token1]
    
    # Calculate the similarity between the current token and all other tokens
    for token2 in model.wv.index_to_key:
        similarity_score = model.wv.similarity(token1, token2)
        similarity_scores.append(similarity_score)
    
    # Append the results to the DataFrame
    similarity_df = similarity_df.append(pd.Series(similarity_scores, index=similarity_df.columns), ignore_index=True)


```
